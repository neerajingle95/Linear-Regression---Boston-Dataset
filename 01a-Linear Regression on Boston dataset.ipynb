### X and y Arrays

```python
boston_df.columns
```

```python
X = boston_df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD',
       'TAX', 'PTRATIO', 'B', 'LSTAT']]
X.head()
```

```python
y = boston_df['MEDV']
y.head()
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

### Creating and Training the Model

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

### Model Evaluation

```python
print(lm.intercept_[0])
```

```python
lm.coef_[0]
```

```python
coeff_df = pd.DataFrame(lm.coef_[0],X.columns,columns=['Coefficient'])
coeff_df
```

- **NOX (nitric oxides concentration (parts per 10 million))** impacts the median house prices the most with a coefficient of -16.23.

- Features **AGE (proportion of owner-occupied units built prior to 1940)** and **B (1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town)** impact house prices the least and we can consider removing them during further analysis.

### Predictions from our Model

```python
predictions = lm.predict(X_test)
predictions
```

```python
plt.scatter(y_test,predictions)
```

We can draw a straight line that fits decently on the scatter plot which means our model has done a pretty good job.

### Residual Histogram

```python
sns.distplot(y_test-predictions,bins=50)
```

Normally distributed residuals means that the linear regression model was a correct choice for the data. Note that the mean is around -2 which should be okay in this case.

### Regression Evaluation Metrics

```python
from sklearn import metrics
```

```python
MAE = metrics.mean_absolute_error(y_test, predictions)
MSE = metrics.mean_squared_error(y_test, predictions)
RMSE = np.sqrt(metrics.mean_squared_error(y_test, predictions))
```

```python
print('MAE:', MAE)
print('MSE:', MSE)
print('RMSE:', RMSE)
```

```python
boston_df['MEDV']
```

```python
mean_price = boston_df['MEDV'].mean()
mean_price
```

```python
RMSE/mean_price['MEDV']*100
```

The RMSE is around 23% of the average house prices in Boston. This percentage error cannot be ignored and definitely opens up
possibilities of using a different model or obtaining a more accurate dataset.
